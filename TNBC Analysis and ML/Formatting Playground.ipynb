# Useful Pandas DF Toolkit

#### Series.str.split(pat=None, n=-1, expand=False)
Splits a string Series(col name of df) into substrings.

pat = delimiter specifying how to split (default is ' ').

n = # of splits; None, 0 or -1 will split everything.

expand = True (optional - returns a df with a separate col per split, new col names will be int not string!).

https://www.geeksforgeeks.org/python-pandas-split-strings-into-two-list-columns-using-str-split/

#### df.rename(columns={'Old Col 1': 'New Col 1', 'Old Col 2' : 'New Col 2'}, inplace= True/False) 
Renames a column or row

inplace = modify the df directly (True) / do not modify original df (False)

#### df.insert(loc, column, value)
loc = column location
column = label of the inserted column
value = Scalar, Series, or Array-like (column from different df)

# CCLE Loading

```python
CCLE = pd.read_csv('Edit Data\Edit_CCLE.csv')
print (CCLE.columns)
print ('Shape =', CCLE.shape)
```

```python
# Splitting and Renaming Columns
Split = CCLE["Unnamed: 0"].str.split("_", n = 1, expand = True) #Split the Column Name
CCLE.insert(1, "Sample", Split[0]) 
CCLE.insert(2, "Cancer Type", Split[1])
CCLE.drop("Unnamed: 0", axis = 1, inplace = True)
CCLE.head()
```

# MCLP Loading

```python
MCLP = pd.read_csv('Edit Data\Edit_MCLP.csv')
print (MCLP.columns)
print ('Shape =', MCLP.shape)
```

```python
# Renaming and inserting new columns
MCLP.rename(columns = {'Sample_Name' : 'Sample'}, inplace = True) #rename Sample_Name to Sample
MCLP.insert(1, "Cancer Type", None) #insert Cancer_Type Column with None values
MCLP.head()
```

```python
# Deesignating Cancer Types based on CCLE Classifications
sMCLP = len(MCLP)
sCCLE = len(CCLE)

# Nested code below iterates through the MCLP Sample column and compares it to every the CCLE Sample column
# If the values match, the MCLP Cancer Type will be replaced with the CCLE Cancer Type
for i in range(sMCLP): 
    for j in range(sCCLE):
        if MCLP.iloc[i, 0] == CCLE.iloc[j, 0]:
            MCLP.iloc[i, 1] = CCLE.iloc[j, 1]
            break #allows exiting of the inner for loop
        else:
            MCLP.iloc[i, 1] = "Undefined"
MCLP.head()
```

```python
print (TCGA['Sample_Type'].unique())
TCGA.rename(columns = {'Sample_ID' : 'Sample', 'Cancer_Type' : 'Cancer Type'}, inplace = True) #rename Sample_Name to Sample
TCGA.head(2)
```

# Unifying Datasets
The CCLE dataset will be used to unify the data as it has the lowest number of protein expression data in the RPPA experiment.

```python
# Determining what features(columns) match across all datasets 
a = list(CCLE.columns.values)
b = list(MCLP.columns.values)
c = list(TCGA.columns.values)

def uniformity(x, y, z):
    uniform_list = []
    for i in x:
        #use in function to detect which items are available
        if i in y and i in z:
            uniform_list.append(i)
    return uniform_list     

u = uniformity(a, b, c)
```

```python
uCCLE = CCLE[u].copy(deep = True) #run a deep copy just incase modification of original list could occur.
print ('uCCLE Shape =', uCCLE.shape)
uCCLE.head()
```

```python
uMCLP = MCLP[u].copy(deep = True) #run a deep copy just incase modification of original list could occur.
print ('uMCLP Shape =', uMCLP.shape)
uMCLP.head()
```

```python
uTCGA = TCGA[u].copy(deep = True) #run a deep copy just incase modification of original list could occur.
print ('uTCGA Shape =', uTCGA.shape)
uTCGA.head(3)
```
